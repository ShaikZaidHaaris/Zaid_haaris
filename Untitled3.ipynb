```python
from sympy import symbols, solve
from scipy.optimize import minimize 
from scipy import optimize
import matplotlib.pyplot as plt


import numpy as np
import math 

#World Solar Challenge is along a 3000 km from Darwin to Adelaide , Australia . The optimisation
#can be performed at any time during 
# race based on the current location , time and battery status. the optimisation takes into account 
# terrain inclination , car rolling resistance , solar power and weather conditions. October is the
#first month of the wet season across the northern half of the Northern Territory. The heat trough over the 
# north of Western Australia has become well established, bringing northwesterly flow over the northern
# NT. Although conditions are mainly fine over the north there is an increased incidence of shower and 
# storm activity. Across the south, dry east to southeasterly winds prevail, although occasionally low 
# pressure troughs can trigger thunderstorms.



#######################################This is the way I created random datasets that give a constraint 
#to work with 
# from faker import Faker
# import random
# import csv
# fake = Faker()




# def get_p1():
#     return random.randrange(0, 1200)
    
# def get_p2():
#     return random.randrange(0, 800)
    
# def get_p3():
#     return random.randrange(0, 40)

# def get_b():
#     return random.randrange(0, 200)

# def get_q():
#     return random.randrange(0, 15)



# def generate_movie():
#     return [get_p1() , get_p2() , get_p3() , get_b() , get_q()]

# with open('movie_data2.csv', 'w') as csvfile:
#     writer = csv.writer(csvfile)
#     writer.writerow(['p1', 'p2' , 'p3' , 'b' , 'q'])
#     for n in range(1, 2000):
#         writer.writerow(generate_movie())




#For every 15 minutes we speed handle and after that we change the speed accordingly
#Firstly we write down the constants of the details of the car
m_Car = 250
c1 = 0.002
g = 9.8
#Next we write the total time for a one day solar race which is from 9 am to 5 pm
time = 8*60*60
#Now we write down the weights of each of the objective function we want to minimize
w1 = 0.5
w2 = 0.5
#These are wind speeds during cloudy weather . Note that we neglect winds in normal conditions .
#We have to predict weather and wind speeds for us to make out a strategy incorporating distubances to 
#our system
q_windy = 1.1
#Here is the Elevation vs Distance data of the Race : 

#Observing the road elevations ,  it is clear that the most difficult day is the second day, while
#the least demanding day is the third day (assuming about 720-820 km /day). The daily consumption of 
#the battery charge should be planned to provide the excess energy to overcome the expected road 
#difficulties
#In this analysis I take into account the velocity profile of a day . Here I take the third day in
#observation . I linearly fit into the curve into distance frame :

#1.800 km to 1450km   y=0 .000682x -352.25

#where y is the inclination  and x is horizontal distance both in metres.
elevation = 0.000682


#####################################SOLAR POWER#######################################################
#This function finds the solar power generated at a given time 
#Note that this is a Gaussian Approximation of a Delayed Differential Equation pertaining
#to prediction of Solar Irradiance
#if weather var is 1 then it means clear sky
#if weather var is 2 then it means partially cloudy which decreases solar power by 45%
#if weather var is 3 then it means heavy cloudy  which decreases solar power by 78%
#Data collected from solving Delayed Differential Equation and curve sketching it 
def s_Find(mins , hrs , weather):
    if(weather==1):
        return 1800*np.exp((-((hrs*60*60 + mins*60)-23200)**2)/112500000)
    elif (weather==2):
        return (55/100)*1800*np.exp((-((hrs*60*60 + mins*60)-23200)**2)/112500000)
    elif (weather == 3):
        return (22/100)*1800*np.exp((-((hrs*60*60 + mins*60)-23200)**2)/112500000)

############################################OBJECTIVE FUNCTION ########################################
#This is the objective function which we want to minimize 
#Note that this contains distance travelled + the difference of target energy consumption and 
#energy consumed at given time t which we wish to minimize
# Between 300 to 400 minutes there is a weather prediction of cloudy sky which at the end of 400 mins
# maybe be high cloudy and windy situation .That is why you could see a dip of speed in the velocity
#vs time graph
def objective(x):
    objective1 = 0
    sum1 = 0
    sum2 = 0
    for i in range(32):
        sum1 = sum1 + x[i] 

    for i in range(32):
        if(i>22 & i<25):
            sum2 = sum2 + (s_Find(15*(1+i) , 1.5 , 2) - (0.045*(x[i]+q_windy)**3 + (c1 + elevation )*m_Car*g*x[i]))
        elif(i>24 & i<26):
            sum2 = sum2 + (s_Find(15*(1+i) , 1.5 , 3 ) - (0.045*(x[i]+ q_windy)**3 + (c1 + elevation )*m_Car*g*x[i]))
        else:
            sum2 = sum2 + (s_Find(15*(1+i) , 1.5 , 1) - (0.045*(x[i])**3 + (c1 + elevation )*m_Car*g*x[i]))
    
    
    return w1*abs((0.72*5000*60*60 + (time/32)*(sum2))) + w2/(sum1*time)

#b is the bound which we want our velocity in . We get this data from the predictive model of the
#dataet 
b = (20,70)
#In between there is a mandatory stop that we can recharge our batteries
b1 = (0,4)
bnds = (b1,b,b,b,b,b,b,b,b,b,b,b,b,b,b,b1,b,b,b,b,b,b,b,b,b,b,b,b,b,b,b,b1)

#This is the test dataset provided
#Note that the test datasets are taken maybe close (not very )to values which are near the predictive values 
#It is so that the minimum does not fall into other local minima so we provide it close to the minima 
#we want.
x0 = [2 ,  34.5 , 34.75 , 46, 48.5 ,  49 ,49.5 , 49.75,  50.2, 5.10 ,51.4,52.8 , 55,58.8,55.5 , 0.5, 49.8 ,49.2  ,47.8, 37.5, 37.4 ,33,22.5 , 22 , 21.5 , 36.5 ,  35.9 , 35.7 ,34.9 , 34.6, 34.5 , 2 ]

########################################OPTIMIZATION/MINIMIZE########################################
#This is the solution i.e the velocities at every 15 minutes 
sol = minimize(objective , x0  , method = 'COBYLA' , bounds=bnds , options={'rhobeg': 0.45 , 'tol':0.001} )
# , options={'eps': 1e-1}
print(sol.x)


# Velocities data
velocities = np.array(sol.x)

time_intervals = np.arange(0, len(velocities) * 15, 15)

# Plotting
plt.plot(time_intervals, velocities)
plt.title('Velocity vs Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Velocity')
plt.grid(True)
plt.show()
power_consumed = (0.045*(velocities+q_windy)**3 + (c1 + elevation )*m_Car*g*velocities)


time_intervals = np.arange(0, len(velocities) * 15, 15)

# Plotting
plt.plot(time_intervals, power_consumed)
plt.title('Power consumed vs Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Power Consumed')
plt.grid(True)
plt.show()

power_given_by_btry = [0] * len(velocities)
for i in range(32):
    if(i>22 & i<25):
        power_given_by_btry[i] = (s_Find(15*(1+i) , 1.5 , 2) - (0.045*(velocities[i]+q_windy)**3 + (c1 + elevation )*m_Car*g*velocities[i]))
    elif(i>24 & i<26):
        power_given_by_btry[i] = (s_Find(15*(1+i) , 1.5 , 3 ) - (0.045*(velocities[i]+ q_windy)**3 + (c1 + elevation )*m_Car*g*velocities[i]))
    else:
        power_given_by_btry[i] = (s_Find(15*(1+i) , 1.5 , 1) - (0.045*(velocities[i])**3 + (c1 + elevation )*m_Car*g*velocities[i]))

remaining_energy = [0] * len(velocities)
rem = 5000

for i in range(32):
    rem = rem + (power_given_by_btry[i])*(1/4)
    remaining_energy[i] = rem

remaining_energy_numpy = np.array(remaining_energy)
plt.plot(time_intervals, remaining_energy_numpy)
plt.title('Remaining energy stored in battery vs Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Energy stored in battery')
plt.grid(True)
plt.show()


total_E_consumption = 0 
for i in range(len(velocities)):
    total_E_consumption =  total_E_consumption + (0.045*(velocities[i]+q_windy)**3 + (c1 + elevation )*m_Car*g*velocities[i])*15*60
print("TOTAL ENERGY CONSUMPTION IS :")
print(total_E_consumption)
    

                
```
